# Nombres falsos

```python
from faker import Faker
import pandas as pd

# Inicializar la librería Faker con localización en México (o la que necesites)
fake = Faker('es_MX')

data = []
for _ in range(300):
    # Generamos el nombre y apellido por separado para evitar prefijos
    nombre = fake.first_name()
    apellido = fake.last_name()


    # Asignación de género para el registro
    genero_val = fake.random_element(elements=('F', 'M'))

    data.append({
        'Nombre': f"{nombre} {apellido}",
        'Genero': genero_val,
        'Edad': fake.random_int(min=18, max=80),
        'Job': fake.job(),
        'Adress': fake.street_address(),
        'City': fake.city(),
        'Phone': fake.phone_number(),
        'CURP': fake.ssn() # Usado como un identificador común, aunque es generado al azar
    })

df_sintetico = pd.DataFrame(data)

df_sintetico
```

# CIUDADES

```python
df_ciudades_id = pd.concat([id_ciudades_norte, id_ciudades_sur], ignore_index=True)
df_ciudades_id
```

```python
import numpy as np

# 1. Extrae las especialidades únicas de tu DataFrame de origen
#    Usamos .values para obtener un array de NumPy con las 60 especialidades
ciudades_disponibles = df_ciudades_id.values

# 2. Determina el número de asignaciones necesarias
num_pacientes = len(df_pacientes) # 150

# 3. Genera la selección aleatoria de especialidades
#    'replace=True' es esencial para que las especialidades se repitan
ciudades_aleatorias = np.random.choice(
    a=ciudades_disponibles,
    size=num_pacientes,
    replace=True
)

# 4. Asigna esta lista a la columna 'especialidad' del DataFrame
df_pacientes["idCiudad"] = ciudades_aleatorias
df_pacientes
```
