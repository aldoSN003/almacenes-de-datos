```python
from faker import Faker
import pandas as pd

# Inicializar la librería Faker con localización en México (o la que necesites)
fake = Faker('es_MX')

data = []
for _ in range(350):
    # Generamos el nombre y apellido por separado para evitar prefijos
    nombre = fake.first_name()
    apellido = fake.last_name()

    # Asignación de género para el registro
    genero_val = fake.random_element(elements=('Femenino', 'Masculino'))

    data.append({
        'Nombre': f"{nombre} {apellido}",
        'Genero': genero_val,
        'Edad': fake.random_int(min=18, max=80),
        'Job': fake.job(),
        'Adress': fake.street_address(),
        'City': fake.city(),
        'Phone': fake.phone_number(),
        'CURP': fake.ssn() # Usado como un identificador común, aunque es generado al azar
    })

df_sintetico = pd.DataFrame(data)

```

```python
names = df_sintetico["Nombre"]



total_rows = len(names)  #  350
midpoint = total_rows // 2 # 175


# Variable 1: Desde el inicio (índice 0) hasta el punto medio (175, exclusivo)
norte_names = names.iloc[0:midpoint]

# Variable 2: Desde el punto medio (índice 175, inclusivo) hasta el final
sur_names = names.iloc[midpoint:]

```

# Datos de los doctores

```python
#-- Tabla Doctores
#CREATE TABLE doctores (
 #   idDoctor VARCHAR(5) PRIMARY KEY,
  #  nombre VARCHAR(30),
   # sueldo INT(11),
    #especialidad VARCHAR(30),
    #idHosp VARCHAR(5),
    #FOREIGN KEY (idHosp) REFERENCES hospitales(idHosp)
#);

doctor_headers = ["idDoctor", "nombre", "sueldo", "especialidad", "idHosp"]
df_doctores_norte = pd.DataFrame(columns=doctor_headers)
df_doctores_sur = pd.DataFrame(columns=doctor_headers)

```

## Zona norte

```python
df_doctores_norte["nombre"] = norte_names
```

```python
df_doctores_norte
```

```python
series_id_hospitales_norte= hospitales_norte["id_hospital"]
series_id_hospitales_norte
```

### ID'S DE HOSPITALES

```python
import numpy as np


# 1. Prepare the IDs
hospital_ids_list = series_id_hospitales_norte.values # Your original 60 unique IDs
num_rows = len(df_doctores_norte) # 150

# 2. Guarantee Initial Appearance (First 60 rows)
#    Assign all 60 unique IDs to the first 60 rows of the 'idHosp' column.
df_doctores_norte["idHosp"].iloc[:60] = hospital_ids_list

# 3. Fill the Remaining Rows (Rows 60 through 149)
#    Calculate the number of remaining rows to fill (150 - 60 = 90)
remaining_size = num_rows - len(hospital_ids_list) # 150 - 60 = 90

# 4. Generate Random Choices for the Remaining Rows
#    Use numpy.random.choice with replacement for the remaining 90 slots
random_filler_ids = np.random.choice(
    a=hospital_ids_list,
    size=remaining_size,
    replace=True
)

# 5. Assign the Random Choices to the Remaining Rows
df_doctores_norte["idHosp"].iloc[60:] = random_filler_ids

```

```python
# Get the counts and immediately call the plot method
import matplotlib.pyplot as plt
# Get the counts, then sort the resulting Series by its index (the ID strings), and plot.
df_doctores_norte["idHosp"].value_counts().sort_index().plot(
    kind='bar',
    figsize=(15, 6),
    title='Frequency of Assigned Hospital IDs (Sorted by ID)',
    xlabel='Hospital ID',
    ylabel='Count'
)

# You will still need to ensure your plot shows correctly:
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Determina el número exacto de IDs necesarios (debería ser 350)
num_doctores = len(df_doctores)

# 2. Genera los identificadores en el formato DR001, DR002, ..., DR350
# Usamos un list comprehension para recorrer los números del 1 al 350
# y f-strings con f":03d" para asegurar que cada número tenga 3 dígitos (ej: 1 -> 001).
id_doctores_350 = [f"DR{i:03d}" for i in range(1, num_doctores + 1)]

# 3. Asigna esta lista a la columna 'idDoctor'
df_doctores["idDoctor"] = id_doctores_350

```
